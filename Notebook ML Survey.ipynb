```python
# Column Mapping 2017

column_mapping = {
'Age':'Age',
'GenderSelect':'Gender',
'Country':'Country',
'FormalEducation':'Education',
'CurrentJobTitleSelect':'Job Title',
'CompensationAmount':'Compensation',  
}

df_2017_new = df_2017.rename(columns= column_mapping)
df_2017_new = df_2017_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2017_new['Year'] = 2017

# Column Mapping 2018

column_mapping = {
'Q2':'Age',
'Q1':'Gender',
'Q3':'Country',
'Q4':'Education',
'Q6':'Job Title',
'Q9':'Compensation',  
}

df_2018_new = df_2018.rename(columns= column_mapping)
df_2018_new = df_2018_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2018_new['Year'] = 2018

# Column Mapping 2019

column_mapping = {'Q1':'Age',
'Q2':'Gender',
'Q3':'Country',
'Q4':'Education',
'Q5':'Job Title',
'Q10':'Compensation',              
}

df_2019_new = df_2019.rename(columns= column_mapping)
df_2019_new = df_2019_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2019_new['Year'] = 2019

# Column Mapping 2020

column_mapping = {'Q1':'Age',
'Q2':'Gender',
'Q3':'Country',
'Q4':'Education',
'Q5':'Job Title',
'Q24':'Compensation',              
}

df_2020_new = df_2020.rename(columns= column_mapping)
df_2020_new = df_2020_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2020_new['Year'] = 2020

# Column Mapping 2021

column_mapping = {'Q1':'Age',
'Q2':'Gender',
'Q3':'Country',
'Q4':'Education',
'Q5':'Job Title',
'Q25':'Compensation',              
}

df_2021_new = df_2021.rename(columns= column_mapping)
df_2021_new = df_2021_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2021_new['Year'] = 2021

# Column Mapping 2022

column_mapping = {'What is your age (# years)?':'Age',
'What is your gender? - Selected Choice':'Gender',
'In which country do you currently reside?':'Country',
'What is the highest level of formal education that you have attained or plan to attain within the next 2 years?':'Education',
'Select the title most similar to your current role (or most recent title if retired): - Selected Choice':'Job Title',
'What is your current yearly compensation (approximate $USD)?':'Compensation',              
}

df_2022_new = df_2022.rename(columns= column_mapping)
df_2022_new = df_2022_new[['Age','Gender','Country','Education','Job Title','Compensation']]
df_2022_new['Year'] = 2022

df = pd.concat([df_2017_new, df_2018_new, df_2019_new, df_2020_new,
                            df_2021_new,df_2022_new], axis=0)
```

```python
df['Country'] = df['Country'].replace(['Hong Kong (S.A.R.)'], 'Hong Kong')
df['Country'] = df['Country'].replace(['United Kingdom'], 'United Kingdom of Great Britain and Northern Ireland')
df['Country'] = df['Country'].replace(['United States'], 'United States of America')
df['Country'] = df['Country'].replace(['Viet Nam'], 'Vietnam')
df['Country'] = df['Country'].replace(['Republic of China'], 'China')
df['Country'] = df['Country'].replace(['Republic of Korea'], 'South Korea')
df['Country'] = df['Country'].replace(['United Kingdom of Great Britain and Northern Ireland'], 'United Kingdom')
df['Country'] = df['Country'].replace(["People 's Republic of China"], 'China')
```

```python
a = df.groupby(['Country','Year']).agg({'Country': ['count']})
a = a.droplevel(0, axis=1)
a = a.reset_index()
a['Country'] = a['Country'].replace(['Iran, Islamic Republic of...'], 'Iran')
a = a.loc[(a['Country'] != "In which country do you currently reside?")
      & (a['Country'] != "I do not wish to disclose my location")
         & (a['Country'] != "Other")]
a = a.reset_index(drop = True)
a['Continent'] = ""
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

a['Continent'] = a['Country'].apply(country_to_continent)
a.rename(columns = {'count':'Count'}, inplace = True)
```

```python
colors = dict(zip(
    ['Africa', 'South America', 'Oceania', 'Europe', 'Asia',
       'North America'],
    ['#adb0ff', '#ffb3ff', '#90d595', '#e48381',
     '#aafbff', '#f7bb5f']
))
group_lk = a.set_index('Country')['Continent'].to_dict()
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
dff = dff[::-1]   # flip values from top to bottom
# pass colors values to `color=`
ax.barh(dff['Country'], dff['Count'], color=[colors[group_lk[x]] for x in dff['Country']])
# iterate over the values to plot labels and values (Tokyo, Asia, 38194.2)
for i, (value, name) in enumerate(zip(dff['Count'], dff['Country'])):
    ax.text(value, i,     name,            ha='right')  # Tokyo: name
    ax.text(value, i-.25, group_lk[name],  ha='right')  # Asia: group name
    ax.text(value, i,     value,           ha='left')   # 38194.2: value
# Add year right middle portion of canvas
ax.text(1, 0.4, current_year, transform=ax.transAxes, size=46, ha='right')
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
def draw_barchart(year):
    dff = a[a['Year'].eq(year)].sort_values(by='Count', ascending=True).tail(10)
    ax.clear()
    ax.barh(dff['Country'], dff['Count'], color=[colors[group_lk[x]] for x in dff['Country']])
    dx = dff['Count'].max() / 200
    for i, (value, name) in enumerate(zip(dff['Count'], dff['Country'])):
        ax.text(value-dx, i,     name,           size=14, weight=600, ha='right', va='bottom')
        ax.text(value-dx, i-.25, group_lk[name], size=10, color='#444444', ha='right', va='baseline')
        ax.text(value+dx, i,     f'{value:,.0f}',  size=14, ha='left',  va='center')
    # ... polished styles
    ax.text(1, 0.4, year, transform=ax.transAxes, color='#777777', size=46, ha='right', weight=800)
    ax.text(0, 1.06, 'Survey Count', transform=ax.transAxes, size=12, color='#777777')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'Kaggle DS & ML Survey',
            transform=ax.transAxes, size=24, weight=600, ha='left')
    ax.text(1, 0, 'by @daniyalarif; credit @jburnmurdoch', transform=ax.transAxes, ha='right',
            color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
    
draw_barchart(2018)
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
fig, ax = plt.subplots(figsize=(15, 8))
animator = animation.FuncAnimation(fig, draw_barchart, frames=range(2017, 2023),interval=5000)
HTML(animator.to_jshtml()) 
# or use animator.to_html5_video() or animator.save()
```
